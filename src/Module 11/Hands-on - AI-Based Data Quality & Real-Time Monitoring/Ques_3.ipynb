## Isolation Forest for Anomaly Detection
**Objective**: Understand and apply the Isolation Forest algorithm to identify anomalies in datasets.

### Task: Anomaly Detection in Sensor Data
**Steps**:
1. Load Dataset
2. Feature Selection
3. Isolation Forest Implementation
4. Plot Results

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Simulated sensor data
np.random.seed(42)
sensor_readings = np.random.normal(loc=20, scale=2, size=1000)  # Normal readings
anomalies = np.random.uniform(low=30, high=40, size=20)         # Anomalous readings
data = np.concatenate([sensor_readings, anomalies])
df = pd.DataFrame({'sensor_value': data})


X = df[['sensor_value']]

# Initialize Isolation Forest
iso_forest = IsolationForest(contamination=0.02, random_state=42)

# Fit and predict
df['anomaly'] = iso_forest.fit_predict(X)


# Plot
plt.figure(figsize=(12, 6))
plt.plot(df['sensor_value'].values, label='Sensor Reading', color='blue')
plt.scatter(df.index[df['anomaly'] == -1], df['sensor_value'][df['anomaly'] == -1],
            color='red', label='Anomaly', marker='x')
plt.title("Anomaly Detection in Sensor Data using Isolation Forest")
plt.xlabel("Time")
plt.ylabel("Sensor Value")
plt.legend()
plt.grid(True)
plt.show()
```
