```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
# commit
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data for height (cm) and weight (kg)
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create a DataFrame
df = pd.DataFrame({
    'Height': heights,
    'Weight': weights
})

# Create the scatter plot using Seaborn
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Height', y='Weight', data=df, color='purple', s=100)

# Add title and labels
plt.title("Scatter Plot of Height vs Weight - Seaborn")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

# Display the plot
plt.tight_layout()
plt.show()
```
