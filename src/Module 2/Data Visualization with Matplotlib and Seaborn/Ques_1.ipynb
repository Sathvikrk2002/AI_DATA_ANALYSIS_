```python
# Creating a Simple Bar Plot with Matplotlib
# Description: Create a bar plot using Matplotlib to visualize the number of sales (in thousands)
# for products: 'A', 'B', 'C', 'D', and 'E' with values [23, 75, 57, 90, 65].

import matplotlib.pyplot as plt

# Product names and corresponding sales (in thousands)
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Create the bar plot
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='skyblue', edgecolor='black')

# Add title and labels
plt.title("Product Sales (in Thousands)")
plt.xlabel("Products")
plt.ylabel("Sales")

# Display the plot
plt.tight_layout()
plt.show()
```
