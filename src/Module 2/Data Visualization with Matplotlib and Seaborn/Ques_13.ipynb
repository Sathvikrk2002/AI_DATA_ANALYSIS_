```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from Seaborn's sample datasets
iris = sns.load_dataset('iris')

# Create a boxplot to visualize the distribution of petal lengths by species
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, x='species', y='petal_length', palette='Set2')

# Add title and labels
plt.title("Distribution of Petal Length by Species", fontsize=16)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
