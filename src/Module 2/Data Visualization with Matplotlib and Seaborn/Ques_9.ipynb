```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.

import matplotlib.pyplot as plt

# Data for the bar plot (Product Sales)
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Data for the scatter plot (Height vs. Weight)
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar plot for product sales
axes[0].bar(products, sales, color='lightcoral', edgecolor='black')
axes[0].set_title("Product Sales (in Thousands)", fontsize=14)
axes[0].set_xlabel("Products", fontsize=12)
axes[0].set_ylabel("Sales", fontsize=12)

# Scatter plot for height vs. weight
axes[1].scatter(heights, weights, color='purple', s=100)
axes[1].set_title("Height vs. Weight", fontsize=14)
axes[1].set_xlabel("Height (cm)", fontsize=12)
axes[1].set_ylabel("Weight (kg)", fontsize=12)

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```
