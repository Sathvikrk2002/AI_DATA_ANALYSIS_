```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.

import matplotlib.pyplot as plt

# Data for the line plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot with customized styles
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='green', linestyle='--', marker='o', markersize=8, linewidth=2)

# Customize the plot's appearance
plt.title("Customized Line Plot", fontsize=16)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True)  # Add gridlines
plt.axhline(0, color='black',linewidth=1)  # Add a horizontal line at y=0
plt.axvline(0, color='black',linewidth=1)  # Add a vertical line at x=0

# Display the plot
plt.tight_layout()
plt.show()
```
