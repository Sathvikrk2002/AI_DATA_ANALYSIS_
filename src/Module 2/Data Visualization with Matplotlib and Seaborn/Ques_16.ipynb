```python
import matplotlib.pyplot as plt
import numpy as np
#
# Data for Bar Plot
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Data for Scatter Plot
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create a figure with 1 row and 2 columns of subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Bar Plot on the first subplot
axs[0].bar(products, sales, color='skyblue')
axs[0].set_title('Product Sales')
axs[0].set_xlabel('Products')
axs[0].set_ylabel('Sales (in thousands)')

# Scatter Plot on the second subplot
axs[1].scatter(heights, weights, color='orange')
axs[1].set_title('Height vs. Weight')
axs[1].set_xlabel('Height (cm)')
axs[1].set_ylabel('Weight (kg)')

# Add some space between the plots
plt.tight_layout()

# Display the combined plots
plt.show()
```
