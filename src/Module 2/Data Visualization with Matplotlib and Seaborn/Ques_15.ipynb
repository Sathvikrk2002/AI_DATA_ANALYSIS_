```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.
#
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from Seaborn's sample datasets
iris = sns.load_dataset('iris')

# Select only the numeric columns (excluding 'species')
iris_numeric = iris.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = iris_numeric.corr()

# Create the heatmap with advanced styling
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, linecolor='black', 
            cbar_kws={'shrink': 0.8}, annot_kws={'size': 12})

# Customize the plot's appearance
plt.title("Correlation Matrix of Iris Dataset", fontsize=16)
plt.tight_layout()

# Display the plot
plt.show()
```
