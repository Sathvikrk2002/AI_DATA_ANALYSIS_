```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
#
import matplotlib.pyplot as plt

# Product names and corresponding sales (in thousands)
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Create the bar plot with customizations
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='lightcoral', edgecolor='black')

# Add title and labels with custom font
plt.title("Product Sales (in Thousands)", fontsize=16, fontweight='bold', family='serif')
plt.xlabel("Products", fontsize=12, fontfamily='serif')
plt.ylabel("Sales", fontsize=12, fontfamily='serif')

# Add a grid with custom styling
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5, color='gray')

# Display the plot
plt.tight_layout()
plt.show()
```
