```python
# Seaborn Histogram with KDE
# Description: Create a histogram using Seaborn that includes a Kernel Density Estimate (KDE) for marks
# data: [23, 45, 56, 78, 89, 90, 67, 88, 76, 59].
#
import seaborn as sns
import matplotlib.pyplot as plt

# Marks data
marks = [23, 45, 56, 78, 89, 90, 67, 88, 76, 59]

# Create the histogram with KDE using Seaborn
plt.figure(figsize=(8, 5))
sns.histplot(marks, kde=True, color='skyblue', bins=5, edgecolor='black')

# Add title and labels
plt.title("Histogram of Marks with KDE", fontsize=16)
plt.xlabel("Marks", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
