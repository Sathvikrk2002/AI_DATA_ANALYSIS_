```python
# Basic Seaborn Bar Plot
# Description: Using Seaborn, plot the same bar plot from Question 1 to compare the sales data.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data for products and sales
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Create a DataFrame
df = pd.DataFrame({
    'Product': products,
    'Sales': sales
})

# Create the bar plot using Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Product', y='Sales', data=df, palette='Blues')

# Add title and labels
plt.title("Product Sales (in Thousands) - Seaborn")
plt.xlabel("Products")
plt.ylabel("Sales")

# Display the plot
plt.tight_layout()
plt.show()
```
