```python
# Title: Advanced Data Visualization
# Description: Create a pair plot for the dataset to display pairwise relationships.

import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add species as a categorical column
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Create a pair plot
sns.set(style="ticks")
pair_plot = sns.pairplot(df, hue='species', corner=True, palette='Set2')

# Show the plot
pair_plot.fig.suptitle("Pair Plot of Iris Dataset", y=1.02)
pair_plot.fig.tight_layout()
pair_plot.fig.subplots_adjust(top=0.95)
```
