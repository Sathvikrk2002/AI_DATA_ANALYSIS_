```python
# Install required libraries (run in terminal or notebook, not Python script):
# !pip install numpy pandas matplotlib seaborn scikit-learn

# Import libraries to verify installation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

print("Libraries imported successfully!")
# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Display first few rows
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Summary of the dataset
print("\nDataset summary:")
print(df.describe(include='all'))
# Convert a column to NumPy array and perform mean and sum
sepal_lengths = df['sepal length (cm)'].values
print("\nMean of Sepal Length:", np.mean(sepal_lengths))
print("Sum of Sepal Length:", np.sum(sepal_lengths))

# Create a NumPy array and calculate variance and standard deviation
arr = np.array([10, 20, 30, 40, 50])
print("\nVariance:", np.var(arr))
print("Standard Deviation:", np.std(arr))

# Use NumPy to filter values greater than 5
filtered = sepal_lengths[sepal_lengths > 5]
print("\nSepal lengths > 5:", filtered)
# Introduce some missing values for demonstration
df.loc[0, 'sepal length (cm)'] = np.nan
df.loc[4, 'petal width (cm)'] = np.nan

# Handle missing data by filling with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Create a new column: petal length to width ratio
df['petal_ratio'] = df['petal length (cm)'] / df['petal width (cm)']

# Group by species and compute the average
grouped_avg = df.groupby('species').mean(numeric_only=True)
print("\nGrouped average by species:")
print(grouped_avg)
# Matplotlib basic plot
plt.figure(figsize=(6, 4))
plt.plot(df['sepal length (cm)'], label='Sepal Length')
plt.title("Sepal Length Line Plot")
plt.xlabel("Index")
plt.ylabel("Length (cm)")
plt.legend()
plt.tight_layout()
plt.show()

# Seaborn histogram
plt.figure(figsize=(6, 4))
sns.histplot(df['petal length (cm)'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Petal Length")
plt.tight_layout()
plt.show()

# Box plot
plt.figure(figsize=(6, 4))
sns.boxplot(x='species', y='sepal length (cm)', data=df, palette='Set2')
plt.title("Sepal Length by Species")
plt.tight_layout()
plt.show()
```
