```python
# Title: Load & Examine a Basic Dataset
# Description: Load an example dataset such as the Iris dataset and perform basic exploration.



# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset using Seaborn
iris_df = sns.load_dataset('iris')

# --- Basic Exploration ---

# Display the first few rows of the DataFrame
print("First 5 rows of the Iris DataFrame:")
print(iris_df.head())
# Observation: The output shows the first five rows of the Iris dataset, including the four features (sepal_length, sepal_width, petal_length, petal_width) and the target variable 'species'.

# Get a concise summary of the DataFrame
print("\nInformation about the Iris DataFrame:")
iris_df.info()
# Observation: The summary indicates that the DataFrame has 150 entries and 5 columns. All columns have 150 non-null values, and the data types are float64 for the features and object for the 'species'.

# Display descriptive statistics of the numerical features
print("\nSummary statistics of the Iris DataFrame:")
print(iris_df.describe())
# Observation: The describe() method provides statistical measures like count, mean, standard deviation, minimum, maximum, and quartiles for each numerical feature. This gives us an initial understanding of the distribution and scale of these features.

# Check for missing values
print("\nNumber of missing values per column:")
print(iris_df.isnull().sum())
# Observation: The output shows that there are no missing values (0 NaN values) in any of the columns of the Iris dataset.

# Explore the 'species' column (categorical feature)
print("\nUnique values in the 'species' column:")
print(iris_df['species'].unique())
# Observation: The 'species' column contains three unique categories: 'setosa', 'versicolor', and 'virginica'.

print("\nValue counts for the 'species' column:")
print(iris_df['species'].value_counts())
# Observation: Each of the three species has an equal number of observations (50). This indicates a balanced dataset in terms of the target variable.

# --- Basic Visualizations ---

# Histograms of numerical features
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.histplot(iris_df['sepal_length'], kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
sns.histplot(iris_df['sepal_width'], kde=True)
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
sns.histplot(iris_df['petal_length'], kde=True)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
sns.histplot(iris_df['petal_width'], kde=True)
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
# Observation: The histograms show the distribution of each numerical feature. Petal length and petal width appear to have more distinct distributions, possibly indicating better separation between species compared to sepal length and sepal width.

# Scatter plot of Sepal Length vs. Sepal Width, colored by Species
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris_df)
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
# Observation: The scatter plot of sepal length vs. sepal width shows some overlap between the species, particularly between versicolor and virginica. Setosa seems to be somewhat separated from the other two based on these features.

# Box plots of Petal Length by Species
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_length', data=iris_df)
plt.title('Petal Length Distribution per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
# Observation: The box plots of petal length for each species show clear differences in the median and spread. Setosa has the smallest petal lengths, followed by versicolor, and then virginica, suggesting that petal length might be a good feature for distinguishing between these species.

# Pair plot to visualize pairwise relationships
sns.pairplot(iris_df, hue='species')
plt.title('Pairwise Relationships in Iris Dataset')
plt.show()
# Observation: The pair plot provides a comprehensive view of the relationships between all pairs of features, colored by species. It visually confirms that petal length and petal width are particularly useful in separating the three species. Setosa appears to be well-separated from versicolor and virginica in several feature combinations.

# --- End of Code ---
```
