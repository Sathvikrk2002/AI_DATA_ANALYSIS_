## Read Data from CSV and Calculate DQI

**Description**: Read data from a CSV file, identify errors as missing values, and calculate the DQI.

```python
import pandas as pd

# Step 1: Load your dataset from CSV
# Replace 'your_file.csv' with your actual file name
df = pd.read_csv('sample.csv')

# Step 2: Display the first few rows
print("Sample Data:")
print(df.head())

# Step 3: Calculate completeness (non-null values)
total_cells = df.size
missing_cells = df.isnull().sum().sum()
complete_cells = total_cells - missing_cells

# Step 4: Compute Data Quality Index (DQI)
dqi = (complete_cells / total_cells) * 100

# Step 5: Output
print(f"\nTotal Cells: {total_cells}")
print(f"Missing Cells: {missing_cells}")
print(f"Complete Cells: {complete_cells}")
print(f"\n✅ Data Quality Index (DQI): {dqi:.2f}%")
```

### Visualize Basic DQI with Bar Plot

**Description**: Create a bar plot for DQI and errors in a dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load your dataset
df = pd.read_csv('sample.csv')  # Replace with your CSV file path

# Step 2: Calculate DQI per column
total_rows = len(df)
missing_counts = df.isnull().sum()
dqi_per_column = 100 * (1 - (missing_counts / total_rows))

# Step 3: Plot DQI per column
plt.figure(figsize=(10, 5))
plt.bar(dqi_per_column.index, dqi_per_column.values, color='green')
plt.title("Data Quality Index (DQI) per Column")
plt.ylabel("DQI (%)")
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 4: Plot Missing Value Counts
plt.figure(figsize=(10, 5))
plt.bar(missing_counts.index, missing_counts.values, color='red')
plt.title("Missing Values per Column")
plt.ylabel("Count of Missing Values")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
