### Monitor Data Quality Trends Over Time

**Task 1**: Create a Trends Analysis Report

**Objective**: Understand long-term data quality trends.

**Steps**:
1. Use historical data (or simulate data) to analyze how data quality has changed over time.
2. Calculate trends for the KPIs defined earlier using statistical measures or visual charts.
3. Write a report summarizing your findings, noting any persistent issues or improvements.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Simulate historical monthly KPI data for 12 months
np.random.seed(42)  # for reproducibility

months = pd.date_range(start='2024-06-01', periods=12, freq='M')
data = {
    'Month': months,
    'Accuracy (%)': np.clip(np.random.normal(75, 10, 12), 50, 100),      # mean 75%, std 10%
    'Completeness (%)': np.clip(np.random.normal(85, 5, 12), 70, 100),   # mean 85%, std 5%
    'Timeliness (%)': np.clip(np.random.normal(65, 15, 12), 30, 95),     # mean 65%, std 15%
}

df = pd.DataFrame(data)

# Step 2: Calculate 3-month rolling averages for smoother trends
df['Accuracy Rolling Avg'] = df['Accuracy (%)'].rolling(window=3).mean()
df['Completeness Rolling Avg'] = df['Completeness (%)'].rolling(window=3).mean()
df['Timeliness Rolling Avg'] = df['Timeliness (%)'].rolling(window=3).mean()

# Step 3: Visualize KPIs and trends
plt.figure(figsize=(12, 7))
plt.plot(df['Month'], df['Accuracy (%)'], marker='o', label='Accuracy')
plt.plot(df['Month'], df['Completeness (%)'], marker='o', label='Completeness')
plt.plot(df['Month'], df['Timeliness (%)'], marker='o', label='Timeliness')

plt.plot(df['Month'], df['Accuracy Rolling Avg'], linestyle='--', color='blue', alpha=0.6)
plt.plot(df['Month'], df['Completeness Rolling Avg'], linestyle='--', color='orange', alpha=0.6)
plt.plot(df['Month'], df['Timeliness Rolling Avg'], linestyle='--', color='green', alpha=0.6)

plt.title('Data Quality KPIs Trends Over 12 Months')
plt.xlabel('Month')
plt.ylabel('Percentage (%)')
plt.ylim(0, 110)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Step 4: Summary Report (printed)
print("=== Data Quality Trends Report ===\n")
print(f"Period Analyzed: {df['Month'].min().strftime('%b %Y')} to {df['Month'].max().strftime('%b %Y')}\n")

for kpi in ['Accuracy (%)', 'Completeness (%)', 'Timeliness (%)']:
    start = df[kpi].iloc[0]
    end = df[kpi].iloc[-1]
    trend = "improved" if end > start else "declined" if end < start else "remained stable"
    print(f"- {kpi[:-4]} started at {start:.1f}%, ended at {end:.1f}% and has {trend} over time.")

print("\nObservations:")
print("- Completeness remains consistently high with minor fluctuations.")
print("- Timeliness shows more volatility and needs targeted improvement.")
print("- Accuracy is moderate with slight fluctuations but shows some improvement.")
```

**Task 2**: Evaluate Continuous Improvement Measures

**Objective**: Implement strategic changes based on trend analysis.

**Steps**:
1. Identify patterns or recurring issues from your trend analysis report.
2. Propose three continuous improvement strategies to address these issues.
3. Plan how to implement these strategies and measure their effectiveness over the next cycle.

```python
%pip install pandas matplotlib numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Step 1: Simulate 12 months of KPI data ---
np.random.seed(42)
months = pd.date_range(start='2024-06-01', periods=12, freq='M')

data = {
    'Month': months,
    'Accuracy (%)': np.clip(np.random.normal(60, 10, 12), 40, 85),      # Simulate moderate accuracy
    'Completeness (%)': np.clip(np.random.normal(80, 7, 12), 65, 95),   # High completeness
    'Timeliness (%)': np.clip(np.random.normal(40, 15, 12), 20, 75),    # Low & volatile timeliness
}

df = pd.DataFrame(data)

# Calculate rolling averages for smoother trend lines
df['Accuracy Rolling Avg'] = df['Accuracy (%)'].rolling(window=3).mean()
df['Completeness Rolling Avg'] = df['Completeness (%)'].rolling(window=3).mean()
df['Timeliness Rolling Avg'] = df['Timeliness (%)'].rolling(window=3).mean()

# --- Step 2: Plot KPI trends ---
plt.figure(figsize=(12, 7))
plt.plot(df['Month'], df['Accuracy (%)'], marker='o', label='Accuracy')
plt.plot(df['Month'], df['Completeness (%)'], marker='o', label='Completeness')
plt.plot(df['Month'], df['Timeliness (%)'], marker='o', label='Timeliness')

plt.plot(df['Month'], df['Accuracy Rolling Avg'], linestyle='--', color='blue', alpha=0.6)
plt.plot(df['Month'], df['Completeness Rolling Avg'], linestyle='--', color='orange', alpha=0.6)
plt.plot(df['Month'], df['Timeliness Rolling Avg'], linestyle='--', color='green', alpha=0.6)

plt.title('Data Quality KPI Trends Over 12 Months')
plt.xlabel('Month')
plt.ylabel('Percentage (%)')
plt.ylim(0, 110)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- Step 3: Summarize trend analysis ---

print("\n=== Data Quality Trends Summary ===\n")
for kpi in ['Accuracy (%)', 'Completeness (%)', 'Timeliness (%)']:
    start = df[kpi].iloc[0]
    end = df[kpi].iloc[-1]
    trend = "improved" if end > start else "declined" if end < start else "remained stable"
    print(f"- {kpi[:-4]} started at {start:.1f}%, ended at {end:.1f}%, and has {trend} over time.")

print("\nKey observations:")
print("- Timeliness is low and fluctuates considerably, indicating delays in data processing.")
print("- Accuracy is moderate with small fluctuations, pointing to occasional data entry issues.")
print("- Completeness remains relatively high but has room for further improvement.\n")

# --- Step 4: Recommend improvement strategies based on trends ---

def recommend_improvements(timeliness, accuracy, completeness):
    improvements = []
    if timeliness < 60:
        improvements.append(
            "Implement automated data update reminders and enforce strict data update SLAs to improve timeliness."
        )
    if accuracy < 70:
        improvements.append(
            "Deploy automated data validation rules and error-checking scripts to increase accuracy."
        )
    if completeness < 85:
        improvements.append(
            "Conduct regular data completeness audits and staff training to maintain high completeness."
        )
    if not improvements:
        improvements.append("Current data quality is good; maintain current processes.")

    return improvements

# Take latest KPI values
latest_accuracy = df['Accuracy (%)'].iloc[-1]
latest_completeness = df['Completeness (%)'].iloc[-1]
latest_timeliness = df['Timeliness (%)'].iloc[-1]

improvement_strategies = recommend_improvements(latest_timeliness, latest_accuracy, latest_completeness)

print("=== Recommended Continuous Improvement Measures ===")
for i, strat in enumerate(improvement_strategies, 1):
    print(f"{i}. {strat}")

# --- Step 5: Suggest implementation plan ---

implementation_plan = pd.DataFrame({
    "Strategy": improvement_strategies,
    "Actions": [
        "Set automatic email reminders and dashboard alerts for data update deadlines.",
        "Integrate validation scripts in data entry pipelines; implement real-time error alerts.",
        "Schedule monthly audits and organize quarterly data quality training sessions."
    ][:len(improvement_strategies)],
    "Metrics to Track": [
        "Percentage of on-time data updates (Timeliness KPI)",
        "Number of data entry errors detected (Accuracy KPI)",
        "Completeness percentage from audits (Completeness KPI)"
    ][:len(improvement_strategies)],
    "Timeline": [
        "1-3 months",
        "1-4 months",
        "Quarterly"
    ][:len(improvement_strategies)],
})

print("\n=== Implementation Plan ===")
print(implementation_plan.to_string(index=False))
```
