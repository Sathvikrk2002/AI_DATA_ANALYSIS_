```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Load the sample dataset
df = sns.load_dataset('tips')

# Choose a numeric column to analyze
column_name = 'total_bill'

# Ensure the chosen column is numeric
if pd.api.types.is_numeric_dtype(df[column_name]):
    # Create the histogram with KDE
    plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
    sns.histplot(df[column_name], kde=True, bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {column_name}', fontsize=16)
    plt.xlabel(column_name, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', alpha=0.75) # Add a subtle grid for easier reading
    sns.despine() # Remove top and right spines for a cleaner look
    plt.show()

    # Calculate and display descriptive statistics
    print(f"\nDescriptive Statistics for '{column_name}':")
    print(df[column_name].describe())
    print(f"Skewness of '{column_name}': {skew(df[column_name]):.2f}")
    print(f"Kurtosis of '{column_name}': {kurtosis(df[column_name]):.2f}")

else:
    print(f"Error: '{column_name}' is not a numeric column. Please choose a numeric column for analysis.")
```
