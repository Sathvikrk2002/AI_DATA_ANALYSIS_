```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Step 1: Create sample customer_data similar to what you described
data = {
    'Age': [25, 45, 35, 50, 23, 43, 36],
    'Annual_Income': [50000, 80000, 62000, 90000, 48000, 75000, 68000],
    'Region': ['North', 'South', 'East', 'West', 'East', 'North', 'South']
}
df = pd.DataFrame(data)

print("Original Data:")
print(df)

# Step 2: Apply MinMaxScaler on Age and Annual_Income
scaler = MinMaxScaler()
df[['Age', 'Annual_Income']] = scaler.fit_transform(df[['Age', 'Annual_Income']])

# Step 3: Perform One-Hot Encoding on Region
df_encoded = pd.get_dummies(df, columns=['Region'])

print("\nTransformed Data after Scaling and One-Hot Encoding:")
print(df_encoded)

# Step 4: Verify transformations by showing summary statistics and head of data
print("\nSummary statistics after scaling:")
print(df_encoded.describe())

print("\nFirst few rows:")
print(df_encoded.head())
```
