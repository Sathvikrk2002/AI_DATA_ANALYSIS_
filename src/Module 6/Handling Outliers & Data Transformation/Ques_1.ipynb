```python
import matplotlib.pyplot as plt
import seaborn as sns  # Optional, for prettier boxplots

def visualize_outliers_boxplot(data):
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=data)
    plt.title("Boxplot to Visualize Outliers")
    plt.xlabel("Values")
    plt.show()

# Example usage
dataset = [10, 12, 11, 13, 12, 10, 11, 100]  # 100 is an outlier
visualize_outliers_boxplot(dataset)
```
